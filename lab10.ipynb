```python
import pandas as pd
import numpy as np
```

```python
index = [('California', 2000), ('California', 2010),('California', 2020),
         ('New York', 2000), ('New York', 2010),('New York', 2020),
         ('Texas', 2000), ('Texas', 2010),('Texas', 2020)]
populations = [33871648, 37253956,37253675,
               18976457, 19378102, 19378675,
               20851820, 25145561, 19378102]
pop = pd.Series(populations, index=index)
pop
```

```python
pop[('California', 2010):('Texas', 2000)]
```

```python
pop[[i for i in pop.index if i[1] == 2010]]
```

```python
#PANDA MULTIINDEXING
index = pd.MultiIndex.from_tuples(index)
index
```

```python
pop = pop.reindex(index)
pop
```

```python
pop[:, 2010]
```

```python
pop_df = pop.unstack()
pop_df
```

```python
pop_df.stack()
```

```python
#column
# hierarchical indices and columns
index = pd.MultiIndex.from_product([[2013, 2014], [1, 2]],
                                   names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']],
                                     names=['subject', 'type'])

# mock some data
data = np.round(np.random.randn(4, 6), 1)
data[:, ::2] *= 10
data += 37

# create the DataFrame
health_data = pd.DataFrame(data, index=index, columns=columns)
health_data
```
